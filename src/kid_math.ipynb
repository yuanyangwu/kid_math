```python
import sys
import random
import numpy as np
import pandas as pd
from pandas import DataFrame
import random
from string import Template

LOWER_LIMIT = 1
UPPER_LIMIT = 20

def gen_basic_add(allow_overflow=True):
    while True:
        a = np.random.random_integers(LOWER_LIMIT, UPPER_LIMIT - LOWER_LIMIT)
        result = np.random.random_integers(a, UPPER_LIMIT)
        b = result - a
        if b >= LOWER_LIMIT:
            if allow_overflow:
                if ((a % 10) + (b % 10)) >= 10:
                    break
                #break
            else:
                break
    return (a, b, result)

def gen_add(allow_overflow=True):
    (a, b, result) = gen_basic_add(allow_overflow)
    return "{:d}+{:d}=".format(a, b)

def gen_subtract(allow_overflow=True):
    (a, b, result) = gen_basic_add(allow_overflow)
    return "{:d}-{:d}=".format(result, a)

def gen_larger_than(blank_index):
    (a, b, result) = gen_basic_add()
    if blank_index == 2:
        return "{:2d}比{:2d}大____?".format(result, a)
    elif blank_index == 1:
        return "{:2d}比____大{:2d}?".format(result, a)
    else:
        return "____比{:2d}大{:2d}?".format(a, b)

def gen_less_than(blank_index):
    (a, b, result) = gen_basic_add()
    if blank_index == 2:
        return "{:2d}比{:2d}小____?".format(a, result)
    elif blank_index == 1:
        return "{:2d}比____小{:2d}?".format(a, b)
    else:
        return "____比{:2d}小{:2d}?".format(result, b)

def gen_add_overflow_smart(has_middel_step):
    a = np.random.random_integers(2, 9)
    b = np.random.random_integers(11 - a, 9)
    if has_middel_step:
        return "{:d}+{:d}={:d}+    +    =".format(a, b, a)
    else:
        return "{:d}+{:d}=".format(a, b, a)

add_formatters = [
    """树上有{}只小鸟，又飞来了{}只小鸟，现在有______只小鸟""",
    """爬走了{}只小猴后，树上有{}只小猴，原来有______只小猴""",
    """小花摘了{}个草莓，小强摘了{}个草莓，一共摘了______个草莓""",
    """草地上有{}只小兔，又跳来了{}只小兔，现在有______只小兔""",
    """车站上有{}个人，又走来了{}个人，现在有______个人""",
    """树上飞走了{}只小鸟后，现在有{}只小鸟，原来有______只小鸟""",
    """小花早上吃了{}个葡萄，小花下午吃了{}个葡萄，小花一天吃了______个葡萄""",
    """爷爷买了{}个苹果，奶奶买了{}个苹果，一共买了______个苹果""",
    """树上有{}只小猴，又爬来了{}只小猴，现在有______只小猴""",
    """跳走了{}只小兔后，草地上有{}只小兔，原来有______只小兔""",
    """车站上现在有{}个人，刚刚离开了{}个人，原来有______个人""",
    """小乌龟爬了{}米，又爬了{}米，一共爬了______米"""
    ]

sub_formatters = [
    """车站上原来有$one个人，______个人走来后，现在有$all个人""",
    """小乌龟爬了$one米，还要爬了______米，才能一共爬了$all米""",
    """草地上原来有$one只小兔，跳来了______只小兔后，草地上现在有$all只小兔""",
    """车站上有$all个人，$one个人离开后，现在有______个人""",
    """小花和小强一共摘了$all个草莓，小花摘了$one个草莓，小强摘了______个草莓""",
    """树上飞来了$one只小鸟，现在有$all只小鸟，树上原来有______只小鸟""",
    """车站上原来有$all个人，______个人离开后，现在有$one个人""",
    """草地上跳来了$one只小兔后，现在有$all只小兔，草地上原来有______只小兔""",
    """爷爷奶奶一共买了$all个苹果，爷爷买了$one个苹果，奶奶买了_____个苹果""",
    """树上原来有$one只小鸟，现在有$all只小鸟，刚刚飞来了______只小鸟，""",
    """草地上现在有$all只小兔，刚才跳来了$one只小兔，草地上原来有______只小兔""",
    """树上现在有$all只小鸟，刚刚飞来了$one只小鸟，树上原来有______只小鸟""",
    """车站上现在有$all个人，刚才走来了$one个人，车站上原来有______个人""",
    """树上原来有$all只小鸟，现在有$one只小鸟，刚刚飞走了______只小鸟，""",
    """草地上有$all只小兔，跳走了$one只小兔后，草地上现在有______只小兔""",
    """树上有$all只小鸟，飞走了$one只小鸟后，现在有______只小鸟""",
    """车站上走来了$one个人后，现在有$all个人，车站上原来有______个人""",
    """草地上原来有$all只小兔，跳走了______只小兔后，草地上现在有$one只小兔""",
    """小花一天吃了$all个葡萄，小花早上吃了$one个葡萄，小花下午吃了______个葡萄"""
    ]

def gen_add_question():
    (a, b, result) = gen_basic_add(True)
    return random.choice(add_formatters).format(a, b)

def gen_sub_question():
    (a, b, result) = gen_basic_add(True)
    return Template(random.choice(sub_formatters)).substitute(all=result, one=a)

def gen_fill_blank_question():
    if np.random.random() > 0.2:
        formatters = [
            """______+$one=$all""",
            """$one+______=$all""",
            """$all-______=$one""",
            """$all-$one=______"""
            ]
        (a, b, result) = gen_basic_add(True)
        return Template(random.choice(formatters)).substitute(all=result, one=a)

    formatters = [
        """______-$one=$other"""
        ]
    (a, b, result) = gen_basic_add(True)
    return Template(random.choice(formatters)).substitute(one=a, other=b)

def gen_half_question():
    if np.random.random() > 0.5:
        formatters = [
            """小兔吃掉一半萝卜后，剩下$half个萝卜，原来有______个萝卜""",
            """哥哥吃掉一半草莓后，剩下$half个草莓，原来有______个草莓"""
            ]
        half = np.random.random_integers(1, 10)
        return Template(random.choice(formatters)).substitute(half=half)

    formatters = [
        """2个小兔平分$all萝卜，每个小兔有______个萝卜""",
        """一共有$all个桃子，2个小猴平分桃子，每个猴子吃______个桃子"""
        ]
    all = np.random.random_integers(1, 10) * 2
    return Template(random.choice(formatters)).substitute(all=all)

def gen_half_plus_question():
    formatters = [
        """白兔黑兔分萝卜，白兔吃掉一半$plus个萝卜，黑兔吃$one个萝卜，白兔吃______个，一共有______个""",
        """姐姐妹妹吃草莓，姐姐吃掉一半$plus个草莓，妹妹吃$one个草莓，姐姐吃______个，一共有______个""",
        """白兔黑兔分萝卜，白兔吃$one个萝卜，白兔再$plus个萝卜后，两个兔子就一样多，黑兔吃______个，一共有______个""",
        """白兔黑兔分萝卜，白兔吃掉一半$plus个萝卜，白兔吃$one个萝卜，黑兔吃______个，一共有______个""",
        """姐姐妹妹吃草莓，姐姐吃掉一半$plus个草莓，姐姐吃$one个草莓，妹妹吃______个，一共有______个"""
        ]
    plus = np.random.random_integers(1, 5)
    if np.random.random() > 0.5:
        plus_str = "多" + str(plus)
    else:
        plus_str = "少" + str(plus)
    one = np.random.random_integers(1, 10) + plus
    return Template(random.choice(formatters)).substitute(one=one, plus=plus_str)

def gen_basic_multiple():
    a = np.random.random_integers(2, 9)
    b = np.random.random_integers(2, 9)
    return (a, b, a * b)

def gen_multiple():
    (a, b, result) = gen_basic_multiple()
    return "{:d}x{:d}=".format(a, b)

def gen_divide():
    (a, b, result) = gen_basic_multiple()
    return "{:d}÷{:d}=".format(result, a)

mul_formatters = [
    """$one个小兔子，每个兔子吃$other个萝卜，一共吃______个萝卜""",
    """每张桌子有$other个小朋友，教室里有$one张桌子，一共有______个小朋友""",
    """有$one个猴子，每个猴子摘了$other个桃子，一共摘了______个桃子"""
    ]

div_formatters = [
    """$one个小兔子一共吃$all个萝卜，每个兔子吃______个萝卜""",
    """一共吃$all个萝卜，每个兔子吃到$one个萝卜，有______只兔子""",
    """一共有$all个小朋友，每张桌子有$one个小朋友，教室里有______张桌子""",
    """$all个小朋友平均做到$one张桌子旁边，每张桌子有______个小朋友""",
    """$all个桃子分给$one个猴子，每个猴子有______个桃子"""
    ]

def gen_mul_question():
    (a, b, result) = gen_basic_multiple()
    return Template(random.choice(mul_formatters)).substitute(one=a, other=b)

def gen_div_question():
    (a, b, result) = gen_basic_multiple()
    return Template(random.choice(div_formatters)).substitute(all=result, one=a)

def gen_queue_question():
    formatters = [
        """一队小朋友，小明$begin，$end，一共有______个小朋友"""
        ]
    begin = np.random.random_integers(2, 5)
    if np.random.random() > 0.5:
        begin_str = "左面有" + str(begin) + "个人"
    else:
        begin_str = "从左数第" + str(begin) + "个"
    end = np.random.random_integers(2, 5)
    if np.random.random() > 0.5:
        end_str = "右面有" + str(begin) + "个人"
    else:
        end_str = "从右数第" + str(begin) + "个"
    return Template(random.choice(formatters)).substitute(begin=begin_str, end=end_str)

def find_biggest(count):
    l = np.random.random_integers(0, UPPER_LIMIT, count)
    return ",".join([str(x) for x in l])
    
def fill_series(count, blank_at_end=False):
    delta_max = min(UPPER_LIMIT / count, 5)
    delta = np.random.random_integers(2, delta_max)
    start = np.random.random_integers(0, UPPER_LIMIT - delta * (count - 1))
    if blank_at_end:
        blank = count - 1
    else:
        blank = np.random.random_integers(0, count - 1)
    return ",".join([str(start + x * delta) if blank != x else "" for x in xrange(0, count)])

def gen_compare_2_items(small, big, is_less_than, is_print_operator):
    op = "   "
    if is_less_than:
        if is_print_operator:
            op = " < "
        return str(small) + op + str(big)
    else:
        if is_print_operator:
            op = " > "
        return str(big) + op + str(small)

def gen_compare():
    a = random.sample(range(1, 100), 3)
    d = [(a[0], 'a'), (a[1], 'b'), (a[2], 'c')]
    d.sort(key=lambda (k, v): k)
    
    r = ""
    r = r + gen_compare_2_items(d[0][1], d[1][1], np.random.random() > 0.5, True)
    r = r + " | "
    r = r + gen_compare_2_items(d[1][1], d[2][1], np.random.random() > 0.5, True)
    r = r + " | "
    r = r + gen_compare_2_items(d[0][1], d[2][1], np.random.random() > 0.5, False)
    return r

def gen_sort():
    a = random.sample(range(1, 100), 3)
    d = [(a[0], 'A'), (a[1], 'B'), (a[2], 'C')]
    d.sort(key=lambda (k, v): k)
    
    r = ""
    r = r + gen_compare_2_items(d[0][1], d[1][1], np.random.random() > 0.5, True)
    r = r + " | "
    r = r + gen_compare_2_items(d[1][1], d[2][1], np.random.random() > 0.5, True)
    r = r + " | "
    r = r + gen_compare_2_items(d[0][1], d[2][1], np.random.random() > 0.5, True)
    return [r, "      >     >     "]

def test():
    print [gen_add() for x in xrange(0,3)] + [gen_subtract() for x in xrange(0,3)]
    print [find_biggest(5) for x in xrange(0,3)]
    print [fill_series(5) for x in xrange(0,3)]

def page1():
    FIND_BIGGEST_COUNT = 12
    find_biggest_list = [find_biggest(4) if x % 2 == 0 else ",,," for x in xrange(0, FIND_BIGGEST_COUNT * 2 - 1)]

    FILL_SERIES_COUNT = 12
    fill_series_list = [fill_series(4) if x % 2 == 0 else ",,," for x in xrange(0, FILL_SERIES_COUNT * 2 - 1)]

    ADD_SUB_COUNT = 23
    add_sub_list = [gen_add() if np.random.random() > 0.5 else gen_subtract() for x in xrange(0, ADD_SUB_COUNT)]

    empty_column = ",".join("" for x in xrange(0, 23))
    #print len(find_biggest_list), len(fill_series_list), len(add_sub_list), len(empty_column)

    df = DataFrame({"fill_series": fill_series_list,
                     "empty1": "",
                     "find_biggest": find_biggest_list,
                     "empty2": "",
                     "empty3": "",
                     "add_subtract": add_sub_list},
                    columns=["fill_series", "empty1", "find_biggest", "empty2", "empty3", "add_subtract"])
    csv = df.to_csv(path_or_buf=sys.stdout, index=False, header = False)

    print """猜猜猜,,,,,排大小,,,,,,加加减减"""
    sys.stdout.write(csv.replace('"', ''))

def page2():
    FILL_SERIES_COUNT = 12
    fill_series_list = [fill_series(4) if x % 2 == 0 else ",,," for x in xrange(0, FILL_SERIES_COUNT * 2 - 1)]

    FILL_ADD_OVERFLOW_SMART_COUNT = 12
    add_overflow_smart_list = [gen_add_overflow_smart(True) if x % 2 == 0 else "" for x in xrange(0, FILL_ADD_OVERFLOW_SMART_COUNT * 2 - 1)]

    ADD_SUB_COUNT = 23
    add_sub_list = [gen_add() if np.random.random() > 0.5 else gen_subtract() for x in xrange(0, ADD_SUB_COUNT)]

    empty_column = ",".join("" for x in xrange(0, 23))
    #print len(add_overflow_smart_list), len(fill_series_list), len(add_sub_list), len(empty_column)

    df = DataFrame({"fill_series": fill_series_list,
                     "empty1": "",
                     "add_overflow_smart_list": add_overflow_smart_list,
                     "empty2": "",
                     "add_subtract": add_sub_list},
                    columns=["fill_series", "empty1", "add_overflow_smart_list", "empty2", "add_subtract"])
    csv = df.to_csv(path_or_buf=sys.stdout, index=False, header = False)

    print """猜猜猜,,,,,凑10,,加加减减"""
    sys.stdout.write(csv.replace('"', ''))
    
def page3():
    large_less_list = [gen_larger_than(0) for x in xrange(0, 11)]
    large_less_list += [""]
    large_less_list += [gen_larger_than(2) for x in xrange(0, 11)]
    #large_less_list += [gen_less_than(0) for x in xrange(0, 11)]

    FILL_ADD_OVERFLOW_SMART_COUNT = 12
    #add_overflow_smart_list = [gen_add_overflow_smart(True) for x in xrange(0, 11)]
    add_overflow_smart_list = [gen_add_overflow_smart(False) for x in xrange(0, 23)]

    compare_list = [gen_compare() if x % 2 == 0 else "" for x in xrange(0, 23)]
    
    ADD_SUB_COUNT = 23
    add_sub_list = [gen_add() if np.random.random() > 0.5 else gen_subtract() for x in xrange(0, ADD_SUB_COUNT)]

    empty_column = ",".join("" for x in xrange(0, 23))
    #print len(add_overflow_smart_list), len(fill_series_list), len(add_sub_list), len(empty_column)

    df = DataFrame({"large_less_list": large_less_list,
                     "empty1": "",
                     "add_overflow_smart_list": add_overflow_smart_list,
                     "empty2": "",
                     "compare": compare_list},
                    columns=["large_less_list", "empty1", "add_overflow_smart_list", "empty2", "compare"])
    csv = df.to_csv(path_or_buf=sys.stdout, index=False, header = False)

    print """大大小小,,凑10,,谁大谁小"""
    sys.stdout.write(csv)#.replace('"', ''))

def page4():
#    add_list = [gen_add(False) for x in xrange(0, 5)] + [""] + [gen_subtract(False) for x in xrange(0, 5)]
#    add_list = add_list + ["谁大谁小"] + [gen_compare() for x in xrange(0,11)]
    add_overflow_list = [gen_add(True) if x % 2 == 0 else "" for x in xrange(0, 23)]
    sub_overflow_list = [gen_subtract(True) if x % 2 == 0 else "" for x in xrange(0, 23)]

    add_overflow_smart_list = [gen_add_overflow_smart(False) for x in xrange(0, 23)]
    sort_list = [item for x in xrange(0, 11) for item in gen_sort()]
    sort_list.append("")

    fill_series_list = [fill_series(4, True) for x in xrange(0, 23)]

    df = DataFrame({"add_overflow_list": add_overflow_list,
                     "sub_overflow_list": sub_overflow_list,
                     "fill_series_list": fill_series_list},
                    columns=["add_overflow_list", "sub_overflow_list", "fill_series_list"])
    csv = df.to_csv(path_or_buf=sys.stdout, index=False, header = False)

    print """100加,100减,猜猜猜"""
    sys.stdout.write(csv.replace('"', ''))

def page5():
    add1_overflow_list = [gen_add(np.random.random() > 0.5) for x in xrange(0, 24)]
    add2_overflow_list = [gen_add(np.random.random() > 0.5) for x in xrange(0, 24)]
    sub1_overflow_list = [gen_subtract(np.random.random() > 0.5) for x in xrange(0, 24)]
    sub2_overflow_list = [gen_subtract(np.random.random() > 0.5) for x in xrange(0, 24)]

    df = DataFrame({"add1_overflow_list": add1_overflow_list,
                     "add2_overflow_list": add2_overflow_list,
                     "sub1_overflow_list": sub1_overflow_list,
                     "sub2_overflow_list": sub2_overflow_list},
                    columns=["add1_overflow_list", "add2_overflow_list", "sub1_overflow_list", "sub2_overflow_list"])
    csv = df.to_csv(path_or_buf=sys.stdout, index=False, header = False)

    sys.stdout.write(csv.replace('"', ''))

def page6():
    add1_overflow_list = [gen_add(np.random.random() > 0.5) for x in xrange(0, 24)]
    sub1_overflow_list = [gen_subtract(np.random.random() > 0.5) for x in xrange(0, 24)]
    compare_list = [gen_compare() for x in xrange(0, 24)]

    df = DataFrame({"add1_overflow_list": add1_overflow_list,
                     "sub1_overflow_list": sub1_overflow_list,
                     "compare_list": compare_list},
                    columns=["add1_overflow_list", "sub1_overflow_list", "compare_list"])
    csv = df.to_csv(path_or_buf=sys.stdout, index=False, header = False)

    sys.stdout.write(csv.replace('"', ''))

def page7():
    questions = [gen_add_question() if np.random.random() > 0.5 else gen_sub_question() for x in xrange(0, 7)]
    #questions += [gen_half_question() for x in xrange(0, 2)]
    #questions += [gen_half_plus_question() for x in xrange(0, 6)]
    questions += [gen_queue_question() for x in xrange(0, 3)]
    for x in xrange(0, 2):
        sub = ''
        for y in xrange(0, 5):
            sub += gen_subtract(True) + '        '
        questions += [sub]
    for q in questions:
        print q
        
def page8():
    questions = [gen_add_question() if np.random.random() > 0.5 else gen_sub_question() for x in xrange(0, 3)]
    questions += [gen_half_question() for x in xrange(0, 1)]
    questions += [gen_queue_question() for x in xrange(0, 1)]
    for x in xrange(0, 1):
        sub = ''
        for y in xrange(0, 5):
            sub += gen_subtract(True) + '        '
        questions += [sub]
    for x in xrange(0, 1):
        sub = ''
        for y in xrange(0, 5):
            sub += gen_multiple() + '        '
        questions += [sub]
    for x in xrange(0, 1):
        sub = ''
        for y in xrange(0, 5):
            sub += gen_divide() + '        '
        questions += [sub]
    for q in questions:
        print q
        
def page9():
    questions = [gen_add_question() if np.random.random() > 0.5 else gen_sub_question() for x in xrange(0, 3)]
    questions += [gen_mul_question() if np.random.random() > 0.5 else gen_div_question() for x in xrange(0, 3)]
    for x in xrange(0, 1):
        sub = ''
        for y in xrange(0, 5):
            sub += gen_add(True) if np.random.random() > 0.5 else gen_subtract(True)
            sub += '        '
        questions += [sub]
    for x in xrange(0, 1):
        sub = ''
        for y in xrange(0, 5):
            sub += gen_multiple() if np.random.random() > 0.5 else gen_divide()
            sub += '        '
        questions += [sub]
    for q in questions:
        print q
        
def page10():
    questions = [gen_add_question() if np.random.random() > 0.5 else gen_sub_question() for x in xrange(0, 1)]
    questions += [gen_queue_question() for x in xrange(0, 1)]
    questions += [gen_mul_question() if np.random.random() > 0.5 else gen_div_question() for x in xrange(0, 4)]
    for x in xrange(0, 2):
        sub = ''
        for y in xrange(0, 5):
            sub += gen_fill_blank_question()
            sub += '        '
        questions += [sub]
    for q in questions:
        print q
```
